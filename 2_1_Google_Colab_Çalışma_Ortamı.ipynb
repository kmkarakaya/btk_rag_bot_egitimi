import matplotlib.pyplot as plt, Python'da grafik çizmek için kullanılan matplotlib kütüphanesinin pyplot modülünü içe aktarır. Genellikle plt kısa adıyla kullanılır. Bu sayede grafik oluşturma, düzenleme ve gösterme gibi işlevleri plt üzerinden çağırabilirsiniz.

````python
def square_numbers():
  """Calculates the square of numbers from 10 to 100."""
  squares = {}
  for i in range(10, 101):
    squares[i] = i**2
  return squares

# Calculate and print the squares
number_squares = square_numbers()
for number, square in number_squares.items():
  print(f"{number} karesi: {square}")
````

````python
import matplotlib.pyplot as plt

# Sayılar ve kareleri
numbers = list(number_squares.keys())
squares = list(number_squares.values())

# Grafik çizimi
plt.figure(figsize=(10, 6))
plt.plot(numbers, squares, marker='o', linestyle='-')
plt.title('10 ile 100 Arasındaki Sayıların Karesi')
plt.xlabel('Sayı')
plt.ylabel('Karesi')
plt.grid(True)
plt.show()
````
